```
import pandas as pd
from fastai.imports import*
from fastai.structured import*
```

```
df_raw = pd.read_csv('HousePrice/train.csv', low_memory= False)
```

```
df_raw.describe()
```

```
y = df_raw.SalePrice
```

```
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
```

```
train_X = df_raw[predictor_cols]
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
my_model = RandomForestRegressor()
my_model.fit(train_X, y)
```

```
df_test = pd.read_csv('HousePrice/test.csv', low_memory =False)
```

```
test_x = df_test[predictor_cols]
```

```
predicted_prices = my_model.predict(test_x)mu
```

```
my_submission = pd.DataFrame({'Id': df_test.Id, 'SalePrice':predicted_prices})
my_submission.to_csv('HousingPrice(kaggleKernel).csv' ,index=False)
```
