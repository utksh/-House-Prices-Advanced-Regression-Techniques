```
df_raw = pd.read_csv("HousePrice/train.csv", low_memory = False, parse_dates=["YearBuilt","YearRemodAdd"])
```

```
def display_all(df):
  with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000):
    display(df)
```

```
display_all(df_raw.tail().T)
```

```
add_datepart(df_raw, 'YearBuilt')
add_datepart(df_raw,'YearRemodAdd')
```

```
display_all(df_raw.tail()
           )
```

```
os.makedirs('tmp', exist_ok=True)
df_raw.to_feather("tmp/HousePrice-raw")
```

```
df , y, nas, mapper = proc_df(df_raw,'SalePrice', do_scale= True)
```

```
display_all(df.tail())
```

```
m= RandomForestRegressor(n_jobs= -1)
m.fit(df,y)
m.score(df,y)
```

```
def rmse(x,y): return math.sqrt(((x-y)**2).mean())

def print_score(m):
  res = [rmse(m.predict(df),y), m.score(df,y)]
  if hasattr(m, 'oob_score_'): res.append(m.oob_score_)
  print(res)
  
```

```
m = RandomForestRegressor(n_jobs=-1)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor(n_estimators=1, n_jobs=-1, max_depth=3,bootstrap = False)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor(n_estimators= 1, n_jobs=-1, bootstrap = False)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor( n_jobs=-1)
m.fit(df,y)
print_score(m)
```

```
preds = np.stack([t.predict(df) for t in m.estimators_])
preds[:,0], np.mean(preds[:,0]) , y[0]
```

```
preds.shape
```

```
m = RandomForestRegressor(n_estimators= 40, n_jobs= -1)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor(n_estimators= 80, n_jobs= -1)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor(n_estimators= 120, n_jobs= -1)
m.fit(df,y)
print_score(m)
```

```
m = RandomForestRegressor(n_estimators= 140, n_jobs= -1)
m.fit(df,y)
print_score(m)
```

**r^2 does not improves on changing the min_samples_leaf**

```
m = RandomForestRegressor(n_estimators=160, n_jobs=-1,max_features= 0.5, oob_score=True)
m.fit(df,y)
print_score(m)
```

```
%time preds = np.stack([t.predict(df) for t in m.estimators_] )
np.mean(preds[:,0]), np.std(preds[:,0])
```
