```
import pandas as pd
from fastai.imports import*
from fastai.structured import*
from sklearn.ensemble import RandomForestRegressor
```

```
df_train = pd.read_csv('HousePrice/train.csv', low_memory =False)
```

```
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```
cols
```

```
df_final = df_train[cols]
```

```
df_final.drop('SalePrice',axis=1, inplace =True  )
```

```
df_final.head()
```

```
df_final.FullBath.value_counts(dropna=False)
```

```
total = df_final.isnull().sum().sort_values(ascending = False)
percentage = (df_final.isnull().sum()/df_final.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total,percentage], axis=1, keys=['Total','Percent'])
missing_data.head(20)
```

```
df_final.YearBuilt.head()
```

```
y = df_train.SalePrice
```

```
df_final = pd.get_dummies(df_final)
```

```
df_final.head()
```

```
m = RandomForestRegressor(n_estimators= 140 , n_jobs= -1 ,max_features=0.5, oob_score=True)
m.fit(df_final, y)
```

```
df_test = pd.read_csv('HousePrice/test.csv', low_memory=False)
```

```
cols_test = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
       'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt']
```

```
df_test_final = df_test[cols_test] 
```

```
df_test_final.head()
```

```
total = df_test_final.isnull().sum().sort_values(ascending=False)
percentage = (df_test_final.isnull().sum()/df_test_final.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total,percentage], axis =1 , keys =['Total','Percent'])
missing_data.head(20)
```

```
predicted_price = m.predict(df_test_final)
```
