```
from fastai.imports import*
from fastai.structured import*
import xgboost as xgb
from pandas_summary import DataFrameSummary
```

```
import os
os.chdir('drive/AI/HousePrice')
```

```
df_raw = pd.read_csv('train.csv', low_memory= False)
```

```
y = df_raw.SalePrice
```

```
predictor_Cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
```

```
train_X = df_raw[predictor_Cols]
```

```
model = xgb.XGBClassifier()
```

```
param_dist = {"max_depth": [10,30,50],
              "min_child_weight" : [1,3,6],
              "n_estimators": [200],
              "learning_rate": [0.05, 0.1,0.16],}
```

```
from sklearn.model_selection import GridSearchCV
```

```
grid_search = GridSearchCV(model, param_grid=param_dist, cv = 3, 
                                   verbose=10, n_jobs=-1)
```

```
grid_search.fit(train_X, y)
```

```
model = xgb.XGBClassifier(max_depth=50, min_child_weight=1,  n_estimators=200,\
                          n_jobs=-1 , verbose=1,learning_rate=0.16)
model.fit(train_X,y)
```

```
df_test = pd.read_csv('test.csv', low_memory =False)
test_x = df_test[predictor_Cols]
```

```
predicted_prices = model.predict(test_x)
```

```
my_submission = pd.DataFrame({'Id':df_test.Id,'SalePrice': predicted_prices})
my_submission.to_csv('XGboost.csv', index= False)
```
